```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.flipkart.com/search?q=tv+under+10000&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_9_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_9_na_na_na&as-pos=3&as-type=RECENT&suggestionId=tv+under+10000&requestId=e02d9d4f-970b-40b2-b768-1219b95e676f&as-backfill=on")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
soup
```

```python
all=soup.find_all("div",{"class":"tUxRFH"})
all
```
